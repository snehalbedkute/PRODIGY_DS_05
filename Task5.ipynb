```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Highway-Rail_Grade_Crossing_Accident_Data.csv", low_memory=False)
```

```python
print("Shape:", df.shape)
print("\nMissing Values:\n", df.isnull().sum().sort_values(ascending=False).head(10))
```

```python
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
```

```python
df['Year'] = df['Date'].dt.year
plt.figure(figsize=(10, 4))
sns.countplot(x='Year', data=df, palette='viridis')
plt.title("Accidents by Year")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Accidents by month
df['Month'] = df['Date'].dt.month
plt.figure(figsize=(10, 4))
sns.countplot(x='Month', data=df, palette='coolwarm')
plt.title("Accidents by Month")
plt.tight_layout()
plt.show()
```

```python
# Accidents by time of day (Hour)
df['Time'] = pd.to_datetime(df['Time'], errors='coerce')
df['Hour'] = df['Time'].dt.hour
plt.figure(figsize=(10, 4))
sns.histplot(df['Hour'].dropna(), bins=24, kde=False, color='orange')
plt.title("Accidents by Hour of Day")
plt.xlabel("Hour")
plt.tight_layout()
plt.show()
```

```python
if 'Weather Condition' in df.columns:
    plt.figure(figsize=(12, 5))
    sns.countplot(y='Weather Condition', data=df, order=df['Weather Condition'].value_counts().head(10).index)
    plt.title("Top 10 Weather Conditions during Accidents")
    plt.tight_layout()
    plt.show()
```

```python
if 'Visibility' in df.columns:
    plt.figure(figsize=(8, 4))
    sns.histplot(df['Visibility'].dropna(), bins=20, kde=True)
    plt.title("Visibility During Accidents")
    plt.tight_layout()
    plt.show()
```

```python
if 'Roadway Condition Code' in df.columns:
    plt.figure(figsize=(10, 5))
    sns.countplot(y='Roadway Condition Code', data=df, order=df['Roadway Condition Code'].value_counts().head(10).index)
    plt.title("Top Roadway Conditions")
    plt.tight_layout()
    plt.show()
```

```python
if 'State Name' in df.columns:
    plt.figure(figsize=(14, 5))
    sns.countplot(y='State Name', data=df, order=df['State Name'].value_counts().head(15).index)
    plt.title("Top 15 States by Number of Accidents")
    plt.tight_layout()
    plt.show()
```

```python
if 'Total Killed Form 57' in df.columns:
    plt.figure(figsize=(6, 4))
    sns.histplot(df['Total Killed Form 57'], bins=10, kde=True)
    plt.title("Distribution of Fatalities")
    plt.tight_layout()
    plt.show()
```

```python
if 'Total Injured Form 57' in df.columns:
    plt.figure(figsize=(6, 4))
    sns.histplot(df['Total Injured Form 57'], bins=10, kde=True)
    plt.title("Distribution of Injuries")
    plt.tight_layout()
    plt.show()
```
